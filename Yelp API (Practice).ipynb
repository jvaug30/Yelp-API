```python
import pandas as pd
import numpy as np
import json
with open('C:\\Users\\javaughn\\OneDrive - HORNE LLP\\Desktop\\yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
from yelpapi import YelpAPI
```

```python
# Create an instance with your key
yelp_api = YelpAPI(login['api-key'], timeout_s=5)
yelp_api
```

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='Pleasant View, TN, 37146 ',
                                       term='food')
print(type(search_results))
search_results.keys()
```

```python
search_results['total']
```

```python
biz = pd.DataFrame(search_results['businesses'])
biz.head(2)
```

```python
## how many businesses in our results
len(search_results['businesses'])
```

```python
# add offset to our original api call
search_results = yelp_api.search_query(location='Pleasant View, TN, 37146 ',
                                       term='food',
                                       offset = 20)
```

```python
biz20 = pd.DataFrame(search_results['businesses'])
biz20.head(2)
```

```python
## concatenate the previous results and new results. 
businesses = pd.concat([biz, biz20],
                      ignore_index=True)
display(businesses.head(3), businesses.tail(3))
```
